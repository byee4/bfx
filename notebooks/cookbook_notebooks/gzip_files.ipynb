# Just a small notebook that gzips all files that can be

```python
from qtools import Submitter
import os
from tqdm import tnrange, tqdm_notebook
import glob
```

```python
cmds = []
num_files = 648
progress = tnrange(num_files)
with open('/projects/ps-yeolab3/bay001/fastq_files.txt', 'r') as f:
    for line in f:
        f = os.path.join('/projects/ps-yeolab3/bay001', line.rstrip())
        if os.path.exists(f) and not os.path.exists(f + '.gz') and not os.path.islink(f):
            cmd = 'gzip {}'.format(line.rstrip())
            # ! $cmd
            print(cmd)
        progress.update(1)
```
