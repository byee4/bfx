# Example for how to convert counts matrices to h5ad format.

```python
import pandas as pd
import scanpy as sc
```

```python
counts_file = '/home/bay001/projects/codebase/bfx/data/GSE162170_rna_counts.tsv.gz'
counts = pd.read_csv(counts_file, sep='\t', index_col=0, compression='gzip')
counts.head()
```

```python
metadata_file = '/home/bay001/projects/codebase/bfx/data/GSE162170_rna_cell_metadata.txt.gz'
metadata = pd.read_csv(metadata_file, sep='\t', index_col=0, compression='gzip')
metadata.head()
```

```python
print(counts.shape, metadata.shape)
```

# Need to match index of counts with index of metadata
- MUST transpose counts (counts.T)

```python
adata = sc.AnnData(X = counts.T, obs = metadata)
print(adata)
```

```python
is_spike_in = {}
number_of_spike_ins = 0

for gene_name in adata.var_names:
    if 'ERCC' in gene_name:
        is_spike_in[gene_name] = True # record that we found a spike-in
        number_of_spike_ins += 1 # bump the counter
    else:
        is_spike_in[gene_name] = False # record that this was not a spike-in
        
adata.var['ERCC'] = pd.Series(is_spike_in) # because the index of adata.var and the keys of is_spike_in match, anndata will take care of matching them up
print('found this many spike ins: ', number_of_spike_ins)
```

```python
adata.write('GSE162170_rna.h5ad') ## the h5ad extension is AnnData-specific
```
