# Reads in and parses Eric's annotations

```python
import pandas as pd
```

```python
input_file = '/home/iachaim/scratch/eclip/Muotri/2019_09_23_NOVA1/third_try/Annotation/WT_reproducible_peaks.bed.annotated_proxdist_miRlncRNA'
```

```python
names = [
    'chr','start','end','-log10pval','log2fc','strand','annotation1','annotation2','geneid_annotation','gene_sym_annotation','region_annotation'
]
df = pd.read_csv(input_file, sep='\t', names=names)
df.head()
```

```python
def get_annotation(row):
    try:
        return row['region_annotation'].split('|')[0]
    except Exception as e:
        return 'intergenic'
```

```python
df['region'] = df.apply(get_annotation, axis=1)
df.head()
```
