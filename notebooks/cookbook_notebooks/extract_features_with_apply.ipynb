# This notebook uses apply to parse and extract a substring within each row inside a dataframe

```python
import pandas as pd
import os
import glob
```

```python
input_dir = '/oasis/tscc/scratch/shsathe/projects/jackie_annotate_idr_peaks/'
```

```python
# start with a dataframe with some header. We'll extract the region from the annotation string.
header = ['chrom','start','end','l10p','l2fc','strand','annotation','gene']
df = pd.read_table(os.path.join(input_dir, 'SKBR3_1-SKBR3_2.idr.final.bed.annotated'), names=header)
df.head()
```

```python
def get_region_from_row(row):
    """ Given a row in a dataframe, parse the annotation column and return the region """
    try:
        return row['annotation'].split('|')[0]
    except IndexError as e:  # this may occur if the string doesn't contain '|' char
        return "Intergenic"
        
df['region'] = df.apply(get_region_from_row, axis=1)
df.head()
```
