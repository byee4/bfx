```python
eric = pd.read_table(
    '/home/bay001/projects/fubp1_stefan_20171214/permanent_data/eCLIP-GATK/pipeline_outputs/stefan_paired_idr/CDT_KHSRP_VEH_01.basedon_CDT_KHSRP_VEH_01.peaks.l2inputnormnew.bed.full.compressed2.bed.full.annotated_proxdist',
    names=['chrom','start','end','name','clip','ip','stat1','stat2','cf','enr','l10p','l2fc','annotation','annotation2','geneid']
)
eric.head()
```

```python
def get_region_from_eric(row):
    try:
        return row['annotation2'].split('|')[0].replace('intron','intron500').replace('3ss','proxintron500').replace('5ss','proxintron500').replace(
        'noncoding_proxintron500','proxnoncoding_intron500').replace('noncoding_distintron500','distnoncoding_intron500')
    except Exception:
        return 'intergenic'
    
def get_strand_from_eric(row):
    return row['name'].split(':')[2]

    
eric['region'] = eric.apply(get_region_from_eric, axis=1)
eric['strand'] = eric.apply(get_strand_from_eric, axis=1)

eric.head()
```

```python
merged = pd.merge(brian, eric, how='outer', left_on=['chrom','start','end','strand'], right_on=['chrom','start','end','strand'])
```
