```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
input_file = 'untitled.txt'

x = pd.read_csv(input_file, sep='\t')
x.head()
```

```python
def get_len(row):
    starts = row['exonStarts'].split(',')
    ends = row['exonEnds'].split(',')
    
    assert len(starts) == len(ends)
    running_total = 0
    for i in range(len(ends)):
        if ends[i] != '':
            running_total += (int(ends[i]) - int(starts[i]))
    return running_total

x['len'] = x.apply(get_len, axis=1)
```
