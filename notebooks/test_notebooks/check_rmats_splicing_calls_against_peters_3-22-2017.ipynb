```python
def subset_rmats(f, pvalue, fdr, sep):
    df = pd.read_table(f)
    dfp = df[
        (df['PValue'] < pvalue) & (df['FDR'] < fdr) & (df['IncLevelDifference'] > sep)
    ]
    dfn = df[
        (df['PValue'] < pvalue) & (df['FDR'] < fdr) & (df['IncLevelDifference'] < -sep)
    ]
    return dfp, dfn
```

```python
pos, neg = subset_rmats(rmats, 0.05, 0.05, 0.05)
```
