```python
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
init_notebook_mode(connected=True)
```

```python
pcdata = pd.read_table('/home/bay001/projects/codebase/data/pca.txt',index_col=0)
counts = '/home/bay001/projects/codebase/data/matrix.txt'

df = pd.read_table(counts, index_col=0)
gene_ids = ['ENSG00000118271', 'ENSG00000111640']
def get_color(df, gene_id):
    colors = df.loc[gene_id]
    return np.log2(colors+1)
```

```python
scatter = go.Scatter3d(
    x=pcdata['0'],
    y=pcdata['1'],
    z=pcdata['2'],
    # mode='markers',
    marker=dict(
        size=12,
        color=get_color(df,gene_ids[0]),
        colorscale='Purples',
        opacity=0.5
    ),
)
```

```python
data = [scatter]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)

button_labels =list([
    dict(
        args=[ { scatter.marker['color'], get_color(df,gene_ids[0]) } ],
        label='gene1',
        method='restyle'
    ),
    dict(
        args=[ { scatter.marker['color'], get_color(df,gene_ids[1]) } ],
        label='gene2',
        method='restyle'
    )
])

updatemenus=list([
    dict(
        buttons = button_labels,
        direction = 'left',
        pad = {'r': 0, 't': 10},
        type = 'buttons',
        x = 0.1,
        xanchor = 'left',
        y = 1.0,
        yanchor = 'top',
        bgcolor = 'AAAAAA',
        active = 99,
        bordercolor = '#FFFFFF',
        font = dict(size=11, color='#000000')
    ),   
])

layout['updatemenus'] = updatemenus
```

```python
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='simple-3d-scatter')
```
