```python
def get_random_sample(fn, n=150):
    df = pd.read_table(fn)
    rand_indices = []
    num_events = df.shape[0]
    for i in range(0, n):
        rand_indices.append(random.randint(0, num_events))
    return df.iloc[rand_indices,].reset_index(drop=True)
```
