```python
import pandas as pd
import numpy as np
import os
```

```python
wd = '/home/bay001/projects/codebase/rbp-maps/examples/se'
names = ['chrom','start','end','l10p','l2fc','strand']
bed = pd.read_table(
    os.path.join(
        wd, '204_01.basedon_204_01.peaks.l2inputnormnew.bed.compressed.bed'
    ),
    names=names
)
```

```python
def p3(df):
    return df[(df['l10p']>=3) & (df['l2fc']>=3)]

bedp3 = p3(bed)
```

```python
bedp3['l10p'] = 'peak'
bedp3['l2fc'] = 0

bedp3.head()
```

```python
bedp3.to_csv(
    os.path.join(
        wd, '204_01.basedon_204_01.peaks.l2inputnormnew.bed.compressed.f3p3.bed'
    ),
    sep='\t',
    index=False,
    header=False
)
```
