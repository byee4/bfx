```python
error_count = 0
add_randomers_to = []
demuxed_fastqs_dir = '/home/gpratt/projects/encode/analysis/encode_split'
# demuxed_fastqs_dir = '/projects/ps-yeolab3/encode/analysis/encode_split'
sample_dict = defaultdict(dict)

for _, row in manifest.iterrows():
    if row['inline_1'] == 'none' and row['inline_2'] == 'none':
        demuxed_fq_r1 = glob.glob(
            os.path.join(
                demuxed_fastqs_dir,
                os.path.basename(row['r1']).replace('.fastq.gz','') + '*.unassigned.fastq.gz'
            )
        )
        if len(demuxed_fq_r1) != 1 or len(demuxed_fq_r1) == 0:
            print('{}'.format(
                os.path.basename(row['r1']).replace('.fastq.gz','') + '*.unassigned.fastq.gz'
            ))
            error_count+=1
        else:
            add_randomers_to.append(demuxed_fq_r1[0])
            sample_dict[row['ENCODE_ID'].replace('_A_IN','').replace('_B_IN','')]['fastq_input_r1'] = demuxed_fq_r1[0].replace('.fastq.gz','.randomer.fastq.gz')
            
        demuxed_fq_r2 = glob.glob(
            os.path.join(
                demuxed_fastqs_dir,
                os.path.basename(row['r2']).replace('.fastq.gz','') + '*.unassigned.fastq.gz'
            )
        )
        if len(demuxed_fq_r2) != 1 or len(demuxed_fq_r2) == 0:
            print('{}'.format(
                os.path.basename(row['r2']).replace('.fastq.gz','') + '*.unassigned.fastq.gz'
            ))
            error_count+=1
        else:
            add_randomers_to.append(demuxed_fq_r2[0])
            sample_dict[row['ENCODE_ID'].replace('_A_IN','').replace('_B_IN','')]['fastq_input_r2'] = demuxed_fq_r2[0].replace('.fastq.gz','.randomer.fastq.gz')
            # print('INPUT!',demuxed_fq_r1)
    else:
        ### INLINE BARCODE 1 READ 1
        dmx_A = row['inline_1']
        demuxed_fq_r1 = glob.glob(
            os.path.join(
                demuxed_fastqs_dir,
                os.path.basename(row['r1']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_A)
            )
        )
        if len(demuxed_fq_r1) != 1 or len(demuxed_fq_r1) == 0:
            print('{}'.format(
                os.path.basename(row['r1']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_A)
            ))
            error_count+=1
        else:
            add_randomers_to.append(demuxed_fq_r1[0])
            sample_dict[row['ENCODE_ID'].replace('_IP_','_')]['fastq_inline1_r1'] = demuxed_fq_r1[0].replace('.fastq.gz','.randomer.fastq.gz')
            
        ### INLINE BARCODE 1 READ 2
        demuxed_fq_r2 = glob.glob(
            os.path.join(
                demuxed_fastqs_dir,
                os.path.basename(row['r2']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_A)
            )
        )
        if len(demuxed_fq_r2) != 1 or len(demuxed_fq_r2) == 0:
            print('{}'.format(
                os.path.basename(row['r2']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_A)
            ))
            error_count+=1
        else:
            add_randomers_to.append(demuxed_fq_r2[0])
            sample_dict[row['ENCODE_ID'].replace('_IP_','_')]['fastq_inline1_r2'] = demuxed_fq_r2[0].replace('.fastq.gz','.randomer.fastq.gz')
        
        ### INLINE BARCODE 2 READ 1
        dmx_B = row['inline_2']
        demuxed_fq_r1 = glob.glob(
            os.path.join(
                demuxed_fastqs_dir,
                os.path.basename(row['r1']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_B)
            )
        )
        if len(demuxed_fq_r1) != 1 or len(demuxed_fq_r1) == 0:
            print('{}'.format(
                os.path.basename(row['r1']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_B)
            ))
            error_count+=1
        else:
            add_randomers_to.append(demuxed_fq_r1[0])
            sample_dict[row['ENCODE_ID'].replace('_IP_','_')]['fastq_inline2_r1'] = demuxed_fq_r1[0].replace('.fastq.gz','.randomer.fastq.gz')
            
        ### INLINE BARCODE 2 READ 2
        demuxed_fq_r2 = glob.glob(
            os.path.join(
                demuxed_fastqs_dir,
                os.path.basename(row['r2']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_B)
            )
        )
        if len(demuxed_fq_r2) != 1 or len(demuxed_fq_r2) == 0:
            print('{}'.format(
                os.path.basename(row['r2']).replace('.fastq.gz','') + "*{}*fastq.gz".format(dmx_B)
            ))
            error_count+=1
        else:
            add_randomers_to.append(demuxed_fq_r2[0])
            sample_dict[row['ENCODE_ID'].replace('_IP_','_')]['fastq_inline2_r2'] = demuxed_fq_r2[0].replace('.fastq.gz','.randomer.fastq.gz')
print('total error: {}'.format(error_count))
```

# parse processed eCLIP files

```python
encode_master_loc = '/projects/ps-yeolab3/encode/analysis/encode_master/'
geo_loc = '/projects/ps-yeolab3/bay001/tbos/geo/geneyeo/'
geo_manifest_file = '/projects/ps-yeolab3/bay001/tbos/geo/processed_clip_files.txt'
manifest[manifest['ENCODE_ID'].str.contains('TIAL_IP_01')]
```

```python
# Process bigbed

from collections import defaultdict

def get_checksum(f):
    x = ! md5sum $f
    return x[0].split(' ')[0]

# softlink the encode_master bigbed location to the GEO upload location
progress = tnrange(len(manifest['ENCODE_ID']))
o = open(geo_manifest_file, 'w')

### Process the IP files
for _, row in manifest.iterrows():
    if 'IP' in row['ENCODE_ID'] and 'OG' not in row['ENCODE_ID']:
        if row['ENCODE_ID'].endswith('_01'):
            rep = '01'
        elif row['ENCODE_ID'].endswith('_02'):
            rep = '02'
        else:
            rep = '00'
        filename = glob.glob(os.path.join(encode_master_loc, '{}*.bb'.format(row['ENCODE_ID'])))
        if len(filename) == 1:
            filename = filename[0]
            base_filename = os.path.basename(filename)
            # ! ln -s $filename $geo_loc # comment out we don't need to softlink again
            o.write(
                '{}\t{}\t{}\t{}\t{}\t{}\n'.format(
                    base_filename,
                    'bigBed',
                    get_checksum(filename),
                    row['RBP'],
                    rep,
                    'IP'
                )
            )
            sample_dict[row['ENCODE_ID'].split('_')[0]+ '_' +rep]['ip_bigBed'] = base_filename
        else:
            print(row['ENCODE_ID'], filename)
    progress.update(1)

### process input files
progress = tnrange(len(manifest['Hiseq_file_name']))
for _, row in manifest.iterrows():
    if 'IN' in row['Hiseq_file_name'] and 'OG' not in row['Hiseq_file_name']:
        if row['ENCODE_ID'].endswith('_01'):
            rep = '01'
        elif row['ENCODE_ID'].endswith('_02'):
            rep = '02'
        else:
            rep = '00'
        filename = glob.glob(os.path.join(encode_master_loc, '{}*.bb'.format(row['Hiseq_file_name'])))
        if len(filename) == 1:
            filename = filename[0]
            base_filename = os.path.basename(filename)
            
            # ! ln -s $filename $geo_loc # comment out we don't need to softlink again
            o.write(
                '{}\t{}\t{}\t{}\t{}\t{}\n'.format(
                    base_filename,
                    'bigBed',
                    get_checksum(filename),
                    row['RBP'],
                    rep,
                    'INPUT'
                )
            )
            sample_dict[row['ENCODE_ID'].split('_')[0] + '_' + rep]['input_bigBed'] = base_filename
        else:
            print(row['Hiseq_file_name'], filename)
    progress.update(1)
```

```python
# process bigwigs
compressed_bed_loc = '/projects/ps-yeolab3/bay001/tbos/geo/geneyeo' # we already softlinked these earlier...

progress = tnrange(len(manifest['ENCODE_ID']))

### Process the IP files
for _, row in manifest.iterrows():
    if 'IP' in row['ENCODE_ID'] and 'OG' not in row['ENCODE_ID']:
        if row['ENCODE_ID'].endswith('_01'):
            rep = '01'
        elif row['ENCODE_ID'].endswith('_02'):
            rep = '02'
        else:
            rep = '00'
        filename_pos = glob.glob(os.path.join(compressed_bed_loc, '{}*.norm.pos.bw'.format(row['ENCODE_ID'])))
        filename_neg = glob.glob(os.path.join(compressed_bed_loc, '{}*.norm.neg.bw'.format(row['ENCODE_ID'])))
        if len(filename_pos) == 1:
            filename_pos = filename_pos[0]
            base_filename = os.path.basename(filename_pos)
            o.write(
                '{}\t{}\t{}\t{}\t{}\t{}\n'.format(
                    base_filename,
                    'bigWig',
                    get_checksum(filename_pos),
                    row['RBP'],
                    rep,
                    'IP'
                )
            )
            sample_dict[row['ENCODE_ID'].split('_')[0]+ '_' +rep]['ip_pos_bw'] = base_filename
        else:
            print(row['ENCODE_ID'], filename, 'pos')
        if len(filename_neg) == 1:
            filename_neg = filename_neg[0]
            base_filename = os.path.basename(filename_neg)
            o.write(
                '{}\t{}\t{}\t{}\t{}\t{}\n'.format(
                    base_filename,
                    'bigWig',
                    get_checksum(filename_neg),
                    row['RBP'],
                    rep,
                    'IP'
                )
            )
            sample_dict[row['ENCODE_ID'].split('_')[0]+ '_' +rep]['ip_neg_bw'] = base_filename
        else:
            print(row['ENCODE_ID'], filename, 'neg')
    elif 'IN' in row['Hiseq_file_name'] and 'OG' not in row['Hiseq_file_name']:
        if row['ENCODE_ID'].endswith('_01'):
            rep = '01'
        elif row['ENCODE_ID'].endswith('_02'):
            rep = '02'
        else:
            rep = '00'
        filename_pos = glob.glob(os.path.join(compressed_bed_loc, '{}*.norm.pos.bw'.format(row['Hiseq_file_name'])))
        filename_neg = glob.glob(os.path.join(compressed_bed_loc, '{}*.norm.neg.bw'.format(row['Hiseq_file_name'])))
        if len(filename_pos) == 1:
            filename_pos = filename_pos[0]
            base_filename = os.path.basename(filename_pos)
            o.write(
                '{}\t{}\t{}\t{}\t{}\t{}\n'.format(
                    base_filename,
                    'bigWig',
                    get_checksum(filename_pos),
                    row['RBP'],
                    rep,
                    'INPUT'
                )
            )
            sample_dict[row['ENCODE_ID'].split('_')[0]+ '_' +rep]['input_pos_bw'] = base_filename
        else:
            print(row['Hiseq_file_name'], filename_pos, 'pos')
        if len(filename_neg) == 1:
            filename_neg = filename_neg[0]
            base_filename = os.path.basename(filename_neg)
            o.write(
                '{}\t{}\t{}\t{}\t{}\t{}\n'.format(
                    base_filename,
                    'bigWig',
                    get_checksum(filename_neg),
                    row['RBP'],
                    rep,
                    'INPUT'
                )
            )
            sample_dict[row['ENCODE_ID'].split('_')[0]+ '_' +rep]['input_neg_bw'] = base_filename
        else:
            print(row['Hiseq_file_name'], filename_neg, 'neg')
    progress.update(1)
```
