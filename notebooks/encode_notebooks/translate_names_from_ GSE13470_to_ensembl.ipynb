```python
import pandas as pd
import os
```

```python
# https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE13470
# /home/bay001/projects/encode/analysis/GSE13470/ncb1800-s2.csv
```

```python
df = pd.read_table('/home/bay001/projects/encode/analysis/GSE13470/ncb1800-s2.csv', sep=',').fillna('.')
df.head()
```

```python
refseq2enst = '/home/bay001/projects/encode/analysis/GSE13470/mart_export.txt'
refseq2enst_df = pd.read_table(refseq2enst).fillna('-')
refseq2enst_mrna_df = refseq2enst_df[['Transcript stable ID', 'RefSeq mRNA ID']]
refseq2enst_ncrna_df = refseq2enst_df[['Transcript stable ID', 'RefSeq ncRNA ID']]

refseq2enst_mrna_df.drop_duplicates(inplace=True)
refseq2enst_ncrna_df.drop_duplicates(inplace=True)
refseq2enst_ncrna_df.head()
```

```python
merged = pd.merge(df, refseq2enst_mrna_df, how='left', left_on='Accession #', right_on='RefSeq mRNA ID').fillna('-')
merged = pd.merge(merged, refseq2enst_ncrna_df, how='left', left_on='Accession #', right_on='RefSeq ncRNA ID').fillna('-')
merged.head()
```

```python
enst2embl = pd.read_table('/home/bay001/projects/encode/analysis/GSE13470/ensembl2embl.txt')
enst2embl.head()
```

```python
merged = pd.merge(merged, enst2embl, how='left', left_on='Accession #', right_on='European Nucleotide Archive ID')
merged.head()
```

```python
merged['Transcript stable ID'].fillna(merged['Transcript stable ID_x'], inplace=True)
merged['Transcript stable ID'].fillna(merged['Transcript stable ID_y'], inplace=True)
```

```python
def isEnst(row):
    if row['Sequence Name(s)'].startswith('ENST'):
        return row['Sequence Name(s)']
    else:
        return '-'
    
merged['last'] = merged.apply(isEnst, axis=1)
```

```python
del merged['Transcript stable ID_x']
del merged['Transcript stable ID_y']
del merged['RefSeq mRNA ID']
del merged['RefSeq ncRNA ID']
del merged['European Nucleotide Archive ID']
del merged['last']
```

```python
merged.to_csv('/home/bay001/projects/encode/analysis/GSE13470/supplemental2_withEnst.tsv', sep='\t', index=False)
```
