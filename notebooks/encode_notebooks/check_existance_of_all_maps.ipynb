```python
def get_reps(line):
    prefix = os.path.basename(line).split('.')[0]
    elements = prefix.split('_')
    if len(elements) != 3:
        uid = elements[0]
        rep = elements[1]
        rbp = elements[3]
    else:
        uid = elements[0]
        rep = elements[1]
        rbp = elements[2]
    return rep

def get_prefix(line):
    prefix = os.path.basename(line).split('.')[0]
    elements = prefix.split('_')
    if len(elements) != 3:
        uid = elements[0]
        rep = elements[1]
        rbp = elements[3]
    else:
        uid = elements[0]
        rep = elements[1]
        rbp = elements[2]
    return uid

def get_rbp(line):
    prefix = os.path.basename(line).split('.')[0]
    elements = prefix.split('_')
    if len(elements) != 3:
        uid = elements[0]
        rep = elements[1]
        rbp = elements[3]
    else:
        uid = elements[0]
        rep = elements[1]
        rbp = elements[2]
    return rbp
```

```python
prefix = defaultdict(list)
for i in images:
    prefix[get_prefix(i)].append(get_reps(i))
    
```

```python
directory = '/projects/ps-yeolab3/bay001/maps/current/se_nr_redo_svgs/'
rawdensities = glob.glob(os.path.join(directory, "*MATS*.ip.raw_density.csv"))
len(rawdensities)
```

```python
def check_secondary_id(filename):
    filename = os.path.basename(filename)
    return filename[filename.find('.r2.')+4:filename.find('-')]

for raw in rawdensities:
    rbp = get_rbp(raw)
    uid = get_prefix(raw)
    secondary = check_secondary_id(raw)
    if rbp != secondary:
        
        remove = os.path.join(directory, "{}_*".format(uid))
        print(uid, rbp, secondary, remove)
        # ! rm $remove
```

```python
import pybedtools

def get_skipped(row):
    return '{}:{}-{}'.format(row['chr'],row['exonStart_0base'],row['exonEnd'])
def print_coords(row):
    return pybedtools.create_interval_from_list(
        [row['chr'],row['upstreamES'], row['downstreamEE'], get_skipped(row), str(row['IncLevelDifference']), row['strand']]
    )
    
def get_peak_intervals(row):
    return pybedtools.create_interval_from_list(
        [row['chrom'],row['start'], row['end'], '0', '0', row['strand']]
    )

rbfox2_rmats = '/projects/ps-yeolab3/bay001/maps/current_annotations/se/RBFOX2-BGHLV26-HepG2-SE.MATS.JunctionCountOnly.negative.nr.txt'
df = pd.read_table(rbfox2_rmats)
dfp = pd.read_table(
    '/home/elvannostrand/data/clip/CLIPseq_analysis/ENCODE_CLIPperv2_20161120/204_01.basedon_204_01.peaks.l2inputnormnew.bed.compressed.bed',
    sep='\t',
    names=['chrom','start','end','l10p','l2fc','strand']
)
dfp = dfp[(dfp['l10p']>3) & (dfp['l2fc']>3)]

df.sort_values(by='IncLevelDifference', inplace=True)
# df['coords'] = df.apply(print_coords,axis=1)

mats = []
peaks = []

for col, row in df.iterrows():
    mats.append(print_coords(row))
    
for col, row in dfp.iterrows():
    peaks.append(get_peak_intervals(row))
    
mats = pybedtools.BedTool(mats)
peaks = pybedtools.BedTool(peaks)
```

```python
dx = peaks.intersect(mats, wa=True, wb=True, s=True).to_dataframe().sort_values(by='blockSizes')
dx.columns = [
    'peak_chrom','peak_start','peak_end','peak_name','peak_score','peak_strand','rmats_chrom', 'rmats_regionstart','rmats_regionend','skipped_incl_exon','sepscore','rmats_strand'
]
dx.to_csv('/home/bay001/projects/encode/analysis/rbfox2_overlapping_peaks_and_rmats.txt',sep='\t',index=None)
```
